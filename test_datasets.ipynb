# Finishers Boston Marathon 2015, 2016 & 2017



```python
df21 = pd.read_csv("data/marathon_results_2015.csv")
df21 = df21.drop(df21.columns[0], axis=1)
df22 = pd.read_csv("data/marathon_results_2016.csv")
df23 = pd.read_csv("data/marathon_results_2017.csv")
df23 = df23.drop(df23.columns[0], axis=1)
```

```python
print(df21.columns)
```

```python
#col_name = 'Unnamed: 9' if 'Unnamed: 9' in df.columns else 'Unnamed: 8'
# rename the column to 'Parathletics' if it exists and replace NaN values with 'No'
def check_col(df):
    print(df.shape)
    col_name = 'Parathletics' 
    if col_name in df.columns:
        print(df[col_name].unique())
        #df[col_name] = df[col_name].fillna('No')
    else:
        print(f"Column '{col_name}' not found in the DataFrame.")
        print("Columns in the DataFrame:", df.columns)

print("Checking df21:")
check_col(df21)
print("Checking df22:")
check_col(df22)
print("Checking df23:")
check_col(df23)
```

```python
df21['Year'] = 2015
df22['Year'] = 2016
df23['Year'] = 2017
```

```python
print(df21.columns)
print(df22.columns)
print(df23.columns)
```

```python
df2 = pd.concat([df21, df22, df23], ignore_index=True)

print(df2.head())
print(df2['Year'].value_counts())

pd.DataFrame.to_csv(df2, "data/all_marathons_results.csv", index=False)
```
