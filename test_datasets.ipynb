# Finishers Boston Marathon 2015, 2016 & 2017



```python
def load_datasets(filepath):
    df = pd.read_csv(filepath)
    has_unnamed = 'Unnamed: 0' in df.columns
    if has_unnamed:
        df = df.drop(columns='Unnamed: 0')
    df = df.rename(columns=lambda x: 'Parathletics' if 'Unnamed:' in x else x)
    df['Parathletics'] = df['Parathletics'].fillna('No')
    return df

df21 = load_datasets("data/marathon_results_2015.csv")
print(df21.columns)
df22 = load_datasets("data/marathon_results_2016.csv")
print(df22.columns)
df23 = load_datasets("data/marathon_results_2017.csv")
print(df23.columns)
```

```python
df21['Year'] = 2015
df22['Year'] = 2016
df23['Year'] = 2017
print(df21.columns)
print(df22.columns)
print(df23.columns)
```

```python
df2 = pd.concat([df21, df22, df23], ignore_index=True)
df2.head()
```

```python
df2 = df2.replace('-', np.nan)
df2.isna().sum()
```

```python


#print(df2.head())
print(df2['Year'].value_counts())
print(df2.shape)
print(df2.isna().sum())
pd.DataFrame.to_csv(df2, "data/all_marathons_results.csv", index=False)
    
```
